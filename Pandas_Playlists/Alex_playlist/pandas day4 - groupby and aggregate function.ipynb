# Group by

## importing library

```python
import pandas as pd
```

## importing  dataset

```python
df = pd.read_csv(r"D:\Data Science\archive\winequality_white.csv")
df
```

## group by on single column

```python
groupby_df = df.groupby("quality")
```

```python
groupby_df.mean()
```

## popular aggregate functions

### aggregate function - mean()

```python
df.groupby("quality").mean()
```

### aggregate function - count()

```python
df.groupby("quality").count()
```

### aggregate function - min()

```python
df.groupby("quality").min()
```

### aggregate function - max()

```python
df.groupby("quality").max()
```

### aggregate function - sum()

```python
df.groupby("quality").sum()
```

### agg() function

```python
df.groupby("quality").agg({'fixed acidity' : ['mean', 'max', 'sum', 'min', 'count'], 'volatile acidity' : ['mean', 'max', 'sum', 'min', 'count']})
```

### group by on multiple column

```python
df.groupby(['quality', 'alcohol']).agg({'fixed acidity' : ['mean', 'max', 'sum', 'min', 'count'], 'volatile acidity' : ['mean', 'max', 'sum', 'min', 'count']})
```

### describe() function

```python
df.groupby(['quality', 'alcohol']).describe()
```
