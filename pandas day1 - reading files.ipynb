## changing the header

```python
pd.read_csv('D:/Data Science/archive/winequality_white.csv', header = None)
```

## giving custom column names 

```python
pd.read_csv('D:/Data Science/archive/winequality_white.csv', header= None, names= ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide','density', 'pH', 'sulphates', 'alcohol', 'quality'])
```

### you can use read_csv function for text file but it is not the proper way.<br> to fix this use 'sep=' seperator parameter.

```python
pd.read_csv('D:/Data Science/archive/winequality_white.txt', sep=',')
```
