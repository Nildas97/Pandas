## sorting the values in ascending order according to year wise population

```python
df.sort_values(by = "2022 Population", ascending=True).head(10)
```

### this gives us idea in ascending order :<br> 1. each country's rank, country name, captial, continent. <br> 2. their year wise population from 2022 to 1970. <br> 3. their area (km²), density per (km²), growth rate and world population percentage

## sorting the values in ascending order according to year wise population

```python
df.sort_values(by = "2022 Population", ascending=False).head(10)
```

### this gives us idea in descending order :<br> 1. each country's rank, country name, captial, continent. <br> 2. their year wise population from 2022 to 1970. <br> 3. their area (km²), density per (km²), growth rate and world population percentage

## checking the correlation

```python
df.corr()
```

### highly correlated values have values nearer to 1 <br> lesser correlated values farther from 1

## visualizing the correlation data using heatmap

```python
sns.heatmap(df.corr(), annot=True)

plt.rcParams['figure.figsize'] = (10,6)

plt.show()
```

### the values in the lighter shade represents higher correlation and values with darker shade represents lower correlation.

## grouping and getting the mean for continent as an index

```python
df.groupby('Continent').mean()
```

## taking only year wise population data 

```python
df.groupby('Continent').mean().sort_values(by="2022 Population", ascending=True)
```

```python
df2 = df.groupby('Continent').mean().sort_values(by="2022 Population", ascending=False)
df2
```

## plotting the year wise population data

```python
df2.plot()
```

### data should only focus on year wise population

## transpose

```python
df3 = df2.transpose()
df3
```

## plotting the transposed data

```python
df3.plot()
```
