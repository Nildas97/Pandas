# Data Cleaning

## importing library

```python
import pandas as pd
```

## importing dataset

```python
df = pd.read_excel(r"D:\Data Science\How to\bank_statement\bank_statement.xlsx")
df
```

## dropping unnecessary column

```python
df.drop(columns= ["Instrument Id", "Unnamed: 2","Unnamed: 4", "Unnamed: 5"], inplace=True)
df
```

## splitting narration column

```python
df['Narration'].str.split('/', expand=True)
```

## after splitting from narration naming the new columns

```python
df[["Transaction_Mode", "Transaction_Id", "CR/DR", "Transaction_Receiver", "Receiver_Bank", "Message-1", "Message-2"]] = df['Narration'].str.split('/', expand=True)
df
```

## filling all the NAN values with null values

```python
df = df.fillna("")
df
```

## replacing NA values with null values

```python
df = df.replace("NA", "")
df
```

## dropping the narration column

```python
df.drop(columns="Narration", inplace=True)
df
```

## reindexing the order of the columns

```python
df = df.reindex(columns=["Transaction Date", "Transaction_Mode", "Transaction_Id", "Transaction_Receiver", "Receiver_Bank", "Message-1", "Message-2", "CR/DR", "Debit", "Credit", "Balance"])
df
```

## standardizing the transaction date column

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce')
```

```python
df['Transaction Date']
```

```python
df['Transaction Date'] = df['Transaction Date'].dt.date
```
