# Pandas Visualization

## importing pandas library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## importing dataset and setting the index 

```python
df = pd.read_csv(r"D:\Data Science\PandasYouTubeSeries-main\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

## selecting the style for visualization

```python
print(plt.style.available)
plt.style.use('seaborn-bright')
```

## line plot

```python
df.plot(kind = 'line')
```

## line plot with subplots

```python
df.plot(kind = 'line', subplots = True)
```

## line plot with xlabels, ylabels and title 

```python
df.plot(kind = 'line', title = "Ice Cream Ratings", xlabel = "Daily Ratings", ylabel = "Scores")
```

## bar plot

```python
df.plot(kind = 'bar')
```

## bar plot with xlabel, ylabel and title

```python
df.plot(kind = 'bar', title = "Ice Cream Ratings", xlabel = "Daily Ratings", ylabel = "Scores")
```

## stacked bar plot

```python
df.plot(kind = 'bar', stacked = True)
```

## bar plot for individual feature

```python
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

## bar plot in horizontal plane

```python
df.plot.barh(stacked = True)
```

## stacked box plot

```python
df.plot.box(stacked = True)
```

## box plot

```python
df.boxplot()
```

## area plot with larger figure size

```python
df.plot.area(figsize = (10,5))
```

## density plot

```python
df.plot.density(stacked = True)
```

## histogram 

```python
df.plot.hist(stacked = True)
```

## KDE plot

```python
df.plot.kde(stacked = True)
```

## line plot

```python
df.plot.line(stacked = True)
```

## scatter plot

```python
df.plot.scatter(x = "Texture Rating", y = "Overall Rating",s = 500, stacked = True)
```

## histogram with xlabel and ylabel

```python
df.plot.hist(x = "Texture Rating", y = "Overall Rating", stacked = True)
```

## histogram with 20 intervals

```python
df.plot.hist(bins = 20)
```

## pie chart with larger figure size

```python
df.plot.pie(y = "Flavor Rating", figsize = (10,6))
```
