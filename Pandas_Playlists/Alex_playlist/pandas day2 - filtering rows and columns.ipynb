# Filtering and Ordering

## importing library

```python
import pandas as pd
```

## importing dataset

```python
df = pd.read_csv(r"D:\Data Science\Dataset\data5\Visadataset.csv")
df
```

## filtering column using conditional operators

```python
df[df['case_id'] <= 'EZYV100']
```

 ## filtering data

### filtering using .isin() function

```python
continents = ['Asia', 'Africa']

df[df['continent'].isin(continents)]
```

### filtering using .contains() function

```python
df[df['education_of_employee'].str.contains('High School')]
```

## filtering using .set_index() function

```python
df2 = df.set_index('continent')
df2
```

## filtering using .filter() function

```python
df2.filter(items = ['region_of_employment', 'education_of_employee'])
```

### here we can mention axis value, which determines which axis it is following. <br> axis = 0, means it is row wise. <br> axis = 1, means it is column wise. 

## filtering using .filter() function and axis = 0

```python
df2.filter(items = ['has_job_experience'], axis=0)
```

## filtering using .filter() function and axis = 1

```python
df2.filter(items = ['region_of_employment', 'education_of_employee'], axis=1)
```

## filtering using filter() function & like

### items method in the above code and like method in the below code works similar

```python
df2.filter(like='Africa', axis= 0)
```

## filtering using loc() function - label based 

```python
df2.loc['Europe']
```

## filtering using iloc() function - integer based

```python
df2.iloc[2251]
```

## filtering using .sort_values() for one column in ascending order

```python
df[df['case_id'] <= 'EZYV100'].sort_values(by='case_id', ascending=True)
```

### here, column case_id less than 100 gets ordered in ascending order.

## filtering using .sort_values() for one column in descending order

```python
df[df['case_id'] <= 'EZYV100'].sort_values(by='case_id', ascending=False)
```

### here, column case_id less than 100 gets ordered in descending order.

## filtering using .sort_values() for multiple column in ascending order

```python
df[df['case_id'] < 'EZYV100'].sort_values(by=['continent', 'education_of_employee'], ascending=True)
```

### here, the column capital get ordered by ascending order and Country remain same.

## filtering using .sort_values() for multiple column in ascending order or descending order

```python
df[df['case_id'] < 'EZYV100'].sort_values(by=['continent', 'education_of_employee'], ascending=[False, True])
```

### sorting the columns in ascending or descending depends on which column gets the priority of getting first place. <br> column Continent gets the priority of first place and gets sorted too but column Capital remains same.
